# Exercice 2

* notes (UserID : entier, MovieID : entier, Rating : entier, Timestamp : entier)
* utils (UserID, Gender, Age, Occupation, Zip-code)
* films (MovieID, Title, Genres)


```scala
val ratings = sc.textFile(folder+ "movielens/ratings.dat")
val users = sc.textFile(folder+ "movielens/users.dat")
val movies = sc.textFile(folder+ "movielens/movies.dat")

//ratings.take(5)
val notes = ratings.map(_.split("::")).map(a => (a(0).toInt, a(1).toInt, a(2).toInt, a(3).toInt))
//notes.take(5)

//users.take(5)
val utils = users.map(_.split("::")).map(a => (a(0).toInt, a(1), a(2).toInt, a(3).toInt, a(4).toInt))


//movies.take(5)
val films = movies.map(_.split("::")).map(a => (a(0).toInt, a(1), a(2)))
//films.take(5)

utils.take(5)
```

```scala
val q21 = notes.map(a => (a._1, 1)).reduceByKey((a,b) => a+b)
q21.take(50)
```

```scala
val joinutils = utils.map(a => (a._1, a._5))
val usersZipCounts = q21.join(joinutils)
//val zipCountsDupl = usersZipCounts.map(t => (t._2._1, t._2._2))
//val q22 = zipCountsDupl.reduceByKey(_+_)
//val zipCountsDupl_2 = usersZipCounts.map(t => (t._1, t._2))
usersZipCounts.take(5)

```
